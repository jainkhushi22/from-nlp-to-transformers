```python
import nltk
from nltk.corpus import stopwords
```

```python
sentence="""NoSQL databases are a category of database management systems that differ fundamentally from 
traditional relational databases. They are designed to handle large volumes of unstructured, 
semi-structured, or structured data, providing flexibility, scalability, and high performance. 
The term 'NoSQL' originally meant 'non-SQL' but is now often interpreted as 'Not Only SQL',
 emphasizing that these databases may still support certain SQL-like querying features. 

Key drivers for the rise of NoSQL include the rapid growth of big data, the need for 
horizontal scalability across distributed systems, and the demand for high-speed data access."""
```

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
for i in range(len(sentence)):
    tokens=nltk.word_tokenize(sentence)
    words=[word for word in tokens if word not in set(stopwords.words('english'))]
    tags = nltk.pos_tag(words)
```
