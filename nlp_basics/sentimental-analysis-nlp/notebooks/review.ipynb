```python
data['overall']=data['overall'].apply(lambda x:0 if x<4 else 1)
```

```python
data.head()
```

```python
data['reviewText']=data.reviewText.str.lower()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from bs4 import BeautifulSoup
```

```python

stop_words = set(stopwords.words("english"))

def clean_text(x):
    x = re.sub(r'http\S+|www\S+', '', x)          #remove url
    x = BeautifulSoup(x,'lxml').get_text()       #remove html
    x = re.sub(r'[^a-zA-Z0-9 ]+', '', x)         #special chars
    x = " ".join([w for w in x.split() if w not in stop_words]) #stopwords
    x = " ".join(x.split())                      #extra spaces
    return x

data['reviewText'] = data['reviewText'].apply(clean_text)
```

```python
data.head()
```

```python
#lemmatizer
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
def lemmatizeword(text):
    return " ".join([lemmatizer.lemmatize(word,'v') for word in text.split()])
```

```python
data['reviewText']=data['reviewText'].apply(lambda x: lemmatizeword(x))
```

```python
data.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test=train_test_split(data['reviewText'],data['overall'],test_size=0.20)
```

Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow= bow.transform(X_test).toarray()
```

TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf=TfidfVectorizer()
X_train_Tfidf=Tfidf.fit_transform(X_train).toarray()
X_test_Tfidf= Tfidf.transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
nbmodel_bow=gb.fit(X_train_bow,y_train)
nbmodel_tfidf=gb.fit(X_train_Tfidf,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
y_pred_bow=nbmodel_bow.predict(X_test_bow)
y_pred_tfidf=nbmodel_tfidf.predict(X_test_Tfidf)
```

```python
acc_bow=accuracy_score(y_test,y_pred_bow)
acc_tfidf=accuracy_score(y_test,y_pred_tfidf)
```

```python
print(acc_bow)
print(acc_tfidf)
```

```python
confusion_matrix(y_test,y_pred_bow)
```

```python
confusion_matrix(y_test,y_pred_tfidf)
```

Word2veb

```python
import gensim
import gensim.downloader as api
wv= api.load("word2vec-google-news-300")
```

```python
import numpy as np

def get_vector(text):

    words=text.split()

    valid_words=[word for word in words if word in wv]

    if len(valid_words)==0:
        return np.zeros(300)

    vectors=[wv[word] for word in valid_words]

    return np.mean(vectors,axis=0)
```

```python
X_train_vec = np.vstack(X_train.apply(get_vector))

X_test_vec = np.vstack(X_test.apply(get_vector))
```

```python
model=GaussianNB()
model.fit(X_train_vec,y_train)
```

```python
from sklearn.linear_model import LogisticRegression

model1=LogisticRegression()

model1.fit(X_train_vec,y_train)
```

```python
pred=model.predict(X_test_vec)
```

```python
pred1=model1.predict(X_test_vec)
```

```python
print(accuracy_score(y_test,pred1))
print(accuracy_score(y_test,pred))
```
