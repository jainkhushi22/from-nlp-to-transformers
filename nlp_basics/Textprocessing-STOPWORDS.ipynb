```python
paragraph="""My dear young friends,

I stand before you with a simple belief: the future of a nation is not written in books or policies, but in the dreams of its people. A dream is not what you see in sleep; a dream is what does not let you sleep.

Knowledge is like a flowing river. When it stops moving, it becomes stagnant. When you continue to learn, to question, and to explore, your mind remains alive. Science and technology are not meant to replace humanity, but to uplift it. Their true purpose is to reduce human suffering and increase human happiness.

Failure is not the opposite of success; it is a part of success. Each failure teaches you a lesson, sharpens your character, and prepares you for responsibility. Do not fear failure. Fear only the situation where you stop trying.

Excellence is not an accident. It is a continuous process of learning, discipline, and perseverance. Small goals may give you comfort, but great goals give you courage. When your goal is high, your efforts automatically rise to meet it.

A nation becomes strong when its citizens are strong in values, integrity, and vision. If you want to change the world, start by changing yourself. When one young mind becomes ignited with purpose, it can light up a thousand more.

My message to you is simple: dream big, work hard, stay honest, and serve the nation with humility. The best way to predict the future is to create it."""
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download("stopwords")
```

```python
stopwords.words('english')
```

```python
stemmer =PorterStemmer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
sentences
```

```python
###apply stopwords and then apply stemming
```

```python
stop_words = set(stopwords.words('english'))

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [
        stemmer.stem(word.lower())
        for word in words
        if word.isalpha() and word.lower() not in stop_words
    ]
    sentences[i] = ' '.join(words)
```

```python
sentences
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
```

```python
stop_words = set(stopwords.words('english'))

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [
        snowballstemmer.stem(word.lower())
        for word in words
        if word.isalpha() and word.lower() not in stop_words
    ]
    sentences[i] = ' '.join(words)
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmetizer=WordNetLemmatizer()
```

```python
stop_words = set(stopwords.words('english'))

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [
        lemmetizer.lemmatize(word.lower())
        for word in words
        if word.isalpha() and word.lower() not in stop_words
    ]
    sentences[i] = ' '.join(words)
```
