```python
sentence="""On 15 March 2024, Riya Sharma joined Infosys as a Data Scientist in Bangalore. 
She completed her Bachelor’s degree from IIT Delhi and previously worked at TCS for two years. 
Riya is currently working on an Artificial Intelligence project for a healthcare client based in the United States. 
Last week, she attended the Google Cloud Summit in Mumbai where she met Sundar Pichai. Her current salary package is ₹12 lakh per annum."""
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
```

```python
token=word_tokenize(sentence)
```

```python
tags = nltk.pos_tag(token)
```

```python
nltk.download("Maxent_NE_Chunker")
```

```python
nltk.download("words")
```

```python
nltk.download('maxent_ne_chunker_tab')  
```

```python
nltk.ne_chunk(tags).draw()
```
