```python
corpus= """hello welcome , to Khushi's NLP tutorials.
please do read entire course! to become expert in NLP."""
```

```python
print(corpus)
```

### Tokenization

sentence--->paragraph

```python
from nltk.tokenize import sent_tokenize
```

```python
documents=sent_tokenize(corpus)
documents
```

paragraph--->words, 
sentence--->words

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

using wordpunct_tokenize(in this 's is divide into 2 words)

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

Using TreebankWordTokenizer(in this only last . is included)

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
