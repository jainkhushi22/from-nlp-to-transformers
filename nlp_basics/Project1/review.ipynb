```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from bs4 import BeautifulSoup
```

```python

stop_words = set(stopwords.words("english"))

def clean_text(x):
    x = re.sub(r'http\S+|www\S+', '', x)          #remove url
    x = BeautifulSoup(x,'lxml').get_text()       #remove html
    x = re.sub(r'[^a-zA-Z0-9 ]+', '', x)         #special chars
    x = " ".join([w for w in x.split() if w not in stop_words]) #stopwords
    x = " ".join(x.split())                      #extra spaces
    return x

data['reviewText'] = data['reviewText'].apply(clean_text)
```

```python
#lemmatizer
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
def lemmatizeword(text):
    return " ".join([lemmatizer.lemmatize(word,'v') for word in text.split()])
```

```python
data['reviewText']=data['reviewText'].apply(lambda x: lemmatizeword(x))
```
