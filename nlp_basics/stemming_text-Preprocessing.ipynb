### STEMMING
Stemming is the process of chopping words down to their root form by removing prefixes or suffixes, often without caring if the result is a real word.

### PorterStemmer

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
```

```python
words= ["eating","eats", "eaten","writes","writes","programming","programs","history","finally","finallized"]
```

```python
for word in words:
    print(word+ "--->" +stemming.stem(word))
```

it can not give some word in right form

```python
stemming.stem("congratulations")
```

### Regexpstemmer class

DIY stemmer. You write the rules. You own the chaos.

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stem=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stem.stem('eating')
```

```python
reg_stem.stem('ingeating') #if we remove $ then eat will come
```

### Snowball Stemmer

Porter, but upgraded.

```python
from nltk.stem import SnowballStemmer
```

```python
ss=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"--->"+ss.stem(word))
```

```python
stemming.stem("fairly")
```

```python
ss.stem("fairly") #snowballsstemmer
```
