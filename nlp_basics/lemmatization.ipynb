Wordnet Lemmatizer

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

pos= "n" for nouns, "v" for verbs, "a" for adjectives, "r" for adverbs and "s" for satellite

```python
lemmatizer.lemmatize('going',pos='n')
```

```python
lemmatizer.lemmatize('going',pos='v')
```

```python
words= ["eating","eats", "eaten","writes","writes","programming","programs","history","finally","finallized"]
```

```python
for word in words:
    print(word+ "--->" +lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize('fairly',pos='v')
```

Stemming → quick, dirty, good for rough preprocessing

WordNet Lemmatizer → accurate, meaningful, production-ready
